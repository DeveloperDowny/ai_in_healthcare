```python
import os

# base_dir = r"D:\DSpit\AIH"
base_dir = !pwd
base_dir = base_dir[0]
base_dir = "../"


csv_path = os.path.join(
    base_dir,
    "lab",
    "Diabetic-Patients-Readmission-Prediction",
    "CSV Files",
    "Diabetes_cleaned.csv",
)
csv_path
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load your dataset
df = pd.read_csv(csv_path)
df.head()
```

```python

# ANOVA to compare cholesterol levels across heart disease status
anova_model = ols('time_in_hospital ~ age', data=df).fit()
anova_table = sm.stats.anova_lm(anova_model, typ=2)
print("ANOVA Results:")
print(anova_table)
```
